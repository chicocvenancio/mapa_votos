```python
locais_compacto = locais[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_ZONA',
       'NR_LOCAL_VOTACAO', 'NM_LOCAL_VOTACAO','DS_ENDERECO', 
       'NM_BAIRRO', 'NR_CEP', 'NR_LATITUDE', 'NR_LONGITUDE',]]
```

```python
locais_compacto = locais_compacto.drop_duplicates()
```

```python
votos_com_locais = pandas.merge(brasil, locais_compacto,on=['SG_UF', 'CD_MUNICIPIO', 'NR_LOCAL_VOTACAO', 'NR_ZONA', 'NM_MUNICIPIO'])
```

```python
votos_com_locais.columns
```

```python
votos_por_predio = votos_com_locais.groupby([
       'CD_MUNICIPIO', 'NM_MUNICIPIO','NR_ZONA', 'SG_UF', 'SG_UE','NM_LOCAL_VOTACAO', 'NR_LOCAL_VOTACAO', 'NM_VOTAVEL', 'NR_VOTAVEL', 'DS_ENDERECO', 'NM_BAIRRO', 'NR_CEP',
       'NR_LATITUDE', 'NR_LONGITUDE',])['QT_VOTOS'].sum()
```

```python
votos_por_predio.to_hdf('votos_por_predio.hd5', 'votos_por_predio')
```

```python
votos_por_predio.to_csv('votos_por_predio.csv')
```

```python
def query_predio(predio, zona, cidade, numero):
    votos = votos_por_predio.to_frame().query(
        f'''NR_LOCAL_VOTACAO == {predio} & NR_ZONA == {zona} & CD_MUNICIPIO == {cidade} & NR_VOTAVEL == {numero}''')['QT_VOTOS']
    return votos

def apply_votos_por_predio_candidatio(row, numero=0):
    #print(row)
    return query_predio(row['NR_LOCAL_VOTACAO'], row['NR_ZONA'], row['CD_MUNICIPIO'], numero)
```
