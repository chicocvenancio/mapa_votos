```python
locais = pandas.read_csv('eleitorado_local_votacao_2022.csv', sep=";", encoding="latin_1")
```

```python
locais[['NM_MUNICIPIO', 'CD_TIPO_LOCAL', 'NR_LATITUDE', 'NR_LONGITUDE', 'DS_ENDERECO', 'NM_BAIRRO', 'NR_CEP', 'NR_LOCAL_VOTACAO', 'NM_LOCAL_VOTACAO', 'SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_ZONA',]]
```

```python
brasil = pandas.read_csv('votacao_secao_2022_BR/votacao_secao_2022_BR.csv', sep=";", encoding="latin_1")
```

```python
brasil.columns
```

```python
locais.columns
```

```python
votos_com_locais = pandas.merge(brasil, locais,on=['SG_UF', 'CD_MUNICIPIO', 'NR_LOCAL_VOTACAO'])
```

```python
votos_com_locais.columns
```

```python
votos_por_predio = votos_com_locais.groupby([
       'CD_MUNICIPIO', 'NM_MUNICIPIO_y', 'SG_UF', 'SG_UE', 'NR_LOCAL_VOTACAO', 'NM_VOTAVEL', 'NR_VOTAVEL', 'DS_ENDERECO', 'NM_BAIRRO', 'NR_CEP',
       'NR_LATITUDE', 'NR_LONGITUDE',])['QT_VOTOS'].sum()
```

```python
votos_por_predio.to_hdf('votos_por_predio.hd5', 'votos_por_predio')
```

```python
votos_por_predio.to_csv('votos_por_predio.csv')
```

```python
votos_por_predio.index
```

```python
votos_por_predio_compacto = votos_com_locais.groupby([
       'CD_MUNICIPIO', 'NR_LOCAL_VOTACAO', 'NR_VOTAVEL',
       ])['QT_VOTOS'].sum()
```

```python
votos_por_predio_compacto.to_csv('votos_por_predio_compacto.csv')
```
